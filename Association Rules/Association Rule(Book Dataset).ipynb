# Apriori Algorithm

## 1. Association rules with 10% Support

```python
freq_itemset=apriori(book,min_support=0.1,use_colnames=True,verbose=1)
freq_itemset
```

```python
rules1=association_rules(freq_itemset,metric='support',min_threshold=0.15)
rules1.sort_values('support',ascending=False)#[0:30]
#support.count()
```

```python
rules1=association_rules(freq_itemset,metric='confidence',min_threshold=0.6)#,support_only=False)
rules1.sort_values('confidence',ascending=False)
#confidence.count()
```

```python
lift_rules=association_rules(freq_itemset,metric='lift',min_threshold=2)
lift_rules[lift_rules.lift>=1]
lift_rules.sort_values('lift',ascending=False)#[0:30]
#lift_rules.count()
```

 A leverage value of 0 indicates independence. Range is [-1 to 1]
 
 A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
plt.scatter(rules1['confidence'],rules1['support'],c = 'aqua',s = 100, marker= '*', edgecolors= 'green')
plt.title('confidence vs support')
plt.xlabel('confidence')
plt.ylabel('support')
```
